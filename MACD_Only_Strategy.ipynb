```python
import numpy as np
import pandas as pd
import yfinance as yf
import vectorbt as vbt

# Function to calculate the Moving Average Convergence Divergence (MACD)
def calculate_macd(df, fast_period=12, slow_period=26, signal_period=9):
    """
    Calculate the MACD and Signal Line.
    """
    # Calculate the MACD
    macd = df['Close'].ewm(span=fast_period, min_periods=1, adjust=False).mean() - df['Close'].ewm(span=slow_period, min_periods=1, adjust=False).mean()

    # Calculate the MACD Signal Line
    macd_signal = macd.ewm(span=signal_period, min_periods=1, adjust=False).mean()

    return macd, macd_signal

# Define the stock symbol and time period
symbol = 'BTC-USD'
start_date = '2019-01-01'
end_date = '2025-01-01'

# Download the data
df = yf.download(symbol, start=start_date, end=end_date)
df.columns = ['Close', 'High', 'Low', 'Open', 'Volume']

# Calculate Stochastic Oscillator (SO) and MACD
df['MACD'], df['MACD_Signal'] = calculate_macd(df, fast_period=12, slow_period=26, signal_period=9)

# Filter data for the test period (2020-2025)
df = df[(df.index.year >= 2020) & (df.index.year <= 2025)]

# Define Entry and Exit signals based on Stochastic Oscillator and MACD
df['Entry'] = (df['MACD'] > df['MACD_Signal'])
df['Exit'] = (df['MACD'] < df['MACD_Signal'])

# Backtest using vectorbt
portfolio = vbt.Portfolio.from_signals(
    close=df['Close'],
    entries=df['Entry'],
    exits=df['Exit'],
    init_cash=100_000,
    fees=0.001,
    slippage=0.002,
    freq='D'
)

# Display performance metrics
print(portfolio.stats())

# Plot equity curve
portfolio.plot().show()
```
